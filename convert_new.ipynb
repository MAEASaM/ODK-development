## Merging section

Merging parent tables to grand-children tables

```python
##joining transects_transects_group_unit_ and  transects_transects_group_unit1 (children) to transects (parent)
transects_unit = transects_transects_group_unit.merge(transects_transects_group_unit_, how= 'left', left_on=['_index'], right_on = ['_parent_index']) \
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'}) \
        .merge(transects_transects_group_unit1, how= 'left', left_on=['_index'], right_on = ['_parent_index']) \
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'}) #Lways renaming index and parent index of parent table
transects_unit.head(2)
```

```python
##joining transects_unit (children) to transects (parent)
transect_merge = transects.merge(transects_unit, how= 'left', left_on=['_index'], right_on = ['_parent_index']) \
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'}) 
# transect_merge.head(20)
```

```python
##joining  threat_type_repeat_potential_th (child) to threat_type_repeat (parent)
threat_merge = threat_type_repeat.merge(threat_type_repeat_potential_th, how= 'left', left_on=['_index'], right_on = ['_parent_index']) \
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'}) 
threat_merge.head(2)
```

### Merging child tables to main parent (main)

```python
main_merge = (main.merge(descriptive_type_repeat, how= 'left', left_on=['_index'], right_on = ['_parent_index']) 
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'})
        .merge(functions_repeat, how= 'left', left_on=['_index'], right_on = ['_parent_index'])
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'}) 
        .merge(landowner_repeat_group, how= 'left', left_on=['_index'], right_on = ['_parent_index'])
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'})
        .merge(natural_feature_photos_repeat, how= 'left', left_on=['_index'], right_on = ['_parent_index'])
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'}) 
        .merge(site_name_repeat, how= 'left', left_on=['_index'], right_on = ['_parent_index'])
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'}) 
        .merge(site_sketch_map_repeat, how= 'left', left_on=['_index'], right_on = ['_parent_index'])
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'})
        .merge(threat_merge, how= 'left', left_on=['_index'], right_on = ['_parent_index'])
         .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'}) 
        .merge(transect_merge, how= 'left', left_on=['_index'], right_on = ['_parent_index'])
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'})\
        .merge(transects_summary_transect_phot, how= 'left', left_on=['_index'], right_on = ['_parent_index'])
        .rename(columns={'_index_x':'_index', '_parent_index_x':'_parent_index'})
             )
```
