```python
# concepts_dict
concept_dict_name = "Institutes.xlsx"
```

```python
import pandas as pd

import xlrd

import openpyxl #required for xlrd 2.0.1 and higher

parent = pd.read_excel('odk_data.xlsm') 
# df.to_csv('your_new_name.csv')

concept_df = pd.read_excel(concept_dict_name)


# Read the excel spreadsheet from the server (): 
# Doing join (parent to children) 
# make blank repeated instances
#  (create dictinary)
# Rename columns
# Export only columns that exist in arches
```

```python
transects =  pd.read_excel('odk_data.xlsm', sheet_name = 'transects')

transects
```

```python
#Join the two tables (left join)
merge = parent.merge(transects, how= 'left', left_on=['_index'], right_on = ['_parent_index'])
```

```python
merge
```

```python
# create concept dict
concept_dict = pd.Series(concept_df.institute.values,index=concept_df.institute_coor).to_dict()
```

```python
# replace values using dict
merge['institute_name'] = merge['institute_name'].map(concept_dict).fillna(merge['institute_name'])
```
