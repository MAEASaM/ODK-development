This script shows how to combine a parent table to a child table (table from repeat group questions): work in progress

```python
import pandas as pd

import xlrd

import openpyxl #required for xlrd 2.0.1 and higher

parent = pd.read_excel('odk_data.xlsm') 
# df.to_csv('your_new_name.csv')
```

```python
transects =  pd.read_excel('odk_data.xlsm', sheet_name = 'transects')

transects
```

```python
#Join the two tables (left join)
merge = parent.merge(transects, how= 'left', left_on=['_index'], right_on = ['_parent_index'])
```
